```python
from pydantic import BaseModel, Field


class Calculations(BaseModel):
    """Get the current weather in a given location"""

    total_spending: float = Field(..., description="The total amount we would spend")
    total_waste:float = Field(..., description="The total amount that is wasted")
    total_deficit: float = Field(..., description="The total defecit we have")


llm_with_tools = llm.bind_tools([Calculations])
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            """We need the totalspending and the total waste, and total defecit from the provided data. What I need is the only three thing as output in dict form
               total_spending:XXXX,
               total_waste:XXXX
               deficit:XXXXX
            """,
        ),
        ("human", "{input}"),
    ]
)

chain = prompt | llm_with_tools
X = chain.invoke(
    {
        
        "input": data,
    }
)
```
